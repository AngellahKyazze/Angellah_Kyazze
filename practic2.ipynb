```python
import matplotlib.pyplot as plt
import numpy as np
df = np.random.normal(loc=50, scale=10, size=1000)
# Add labels and title
plt.hist(df, bins=30, density=True)
plt.xlabel("Unit_Cost")
plt.ylabel("Cost")
plt.title("Histogram of Continuous Variable by Category")

# Show the plot
plt.show()
```

```python

import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Generate random data (replace with your actual data)
data = np.random.normal(loc=50, scale=10, size=(1000, 2))  # Create 2D array for Cost and Unit_Cost

# Convert the array to a DataFrame
df = pd.DataFrame(data, columns=['Cost', 'Unit_Cost'])

# Create a scatter plot
sns.scatterplot(x='Cost', y='Unit_Cost', data=df)

# Add labels and title (optional)
plt.xlabel("Unit Cost")
plt.ylabel("Cost")
plt.title("Scatter Plot of Cost vs. Unit Cost")

plt.show()

```
